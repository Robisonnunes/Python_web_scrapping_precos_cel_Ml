```python
import pandas as pd
from datetime import date
import time
import requests
from bs4 import BeautifulSoup

import warnings
warnings.filterwarnings("ignore")

valores = []
modelos = []
urls = 'https://celulares.mercadolivre.com.br/novo/'

for i in range(51,1951,50):
    url = 'https://celulares.mercadolivre.com.br/novo/_Desde_'+str(i)+'_NoIndex_True'
    page = requests.get('https://celulares.mercadolivre.com.br/novo/_Desde_'+str(i)+'_NoIndex_True')
    soup = BeautifulSoup(page.content, 'html.parser')
    precos = soup.find_all('li', class_='ui-search-layout__item')

    for i in range(len(precos)):
        preco = soup.find_all('li', class_='ui-search-layout__item')[i]
        preco1 = preco.find_all('span', class_='price-tag-fraction')[0]
        preco2 = preco1.get_text()
        preco2 = preco2.replace('.', '')
        preco2 = float(preco2)
        valores.append(preco2)

    for i in range(len(precos)):
        preco = soup.find_all('li', class_='ui-search-layout__item')[i]
        title1 = preco.find_all('h2', class_='ui-search-item__title')[0]
        title2 = title1.get_text()
        modelos.append(title2) 
        
celulares_novos = pd.DataFrame({'type':'novo', 'model':modelos, 'price':valores, 'page':urls, 'date':date.today()})                 

valores1 = []
modelos1 = []
urls1 = 'https://celulares.mercadolivre.com.br/usados/'

for i in range(0,1951,50):
    url = 'https://celulares.mercadolivre.com.br/usados/_Desde_'+str(i)+'_NoIndex_True'
    page = requests.get('https://celulares.mercadolivre.com.br/usados/_Desde_'+str(i)+'_NoIndex_True')
    soup = BeautifulSoup(page.content, 'html.parser')
    precos = soup.find_all('li', class_='ui-search-layout__item')

    for i in range(len(precos)):
        preco = soup.find_all('li', class_='ui-search-layout__item')[i]
        preco1 = preco.find_all('span', class_='price-tag-fraction')[0]
        preco2 = preco1.get_text()
        preco2 = preco2.replace('.', '')
        preco2 = float(preco2)
        valores1.append(preco2)

    for i in range(len(precos)):
        preco = soup.find_all('li', class_='ui-search-layout__item')[i]
        title1 = preco.find_all('h2', class_='ui-search-item__title')[0]
        title2 = title1.get_text()
        modelos1.append(title2) 
        
           
celulares_usados = pd.DataFrame({'type':'usado','model':modelos1, 'price':valores1, 'page':urls1, 'date':date.today()})         

celulares = pd.concat([celulares_novos, celulares_usados])
celulares.to_excel('celulares_{}.xls'.format(date.today()), index= False) 
```
